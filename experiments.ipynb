```python
target = "Churn"

categorical_columns = [column for column in df.columns if df[column].dtype=='object' and column != target]
categorical_columns
```

```python
numerical_columns = [column for column in df.columns if column not in categorical_columns and column != target]
numerical_columns
```

```python
X = df.drop(target,axis=1)  # Independent features
y = df[target]              # Dependent feature
```

```python
from pandas.core.groupby.groupby import RandomState
from sklearn.model_selection import train_test_split

X_train,X_text,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
print(len(X_train), len(X_text), len(y_train), len(y_test))
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

```python
from matplotlib.pyplot import step


num_pipeline = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]
)


cat_pipiline = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ]
)
```

```python
preprocessor = ColumnTransformer([
    ('num', num_pipeline, numerical_columns),
    ('cat', cat_pipiline, categorical_columns)
], remainder='drop')
```

```python
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_text)
```

```python
num_features = numerical_columns
cat_features = preprocessor.named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_columns)
all_features = list(num_features) + list(cat_features)
```

```python
X_train_transformed_df = pd.DataFrame(X_train_preprocessed, columns=all_features, index= X_train.index)
```

```python
X_train_transformed_df
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_balanced, y_train_balanced = smote.fit_resample(X_train_preprocessed, y_train)
```

```python
print("Original y_train distribution:\n", y_train.value_counts())
print("Balanced y_train distribution:\n", pd.Series(y_train_balanced).value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, confusion_matrix, precision_recall_curve, average_precision_score
```

```python
from sklearn.metrics import classification_report, roc_auc_score

models = [
    LogisticRegression(max_iter=1000),
    SVC(probability=True),
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    XGBClassifier(use_label_encoder=False, eval_metric='logloss')
]

for model in models:
    model.fit(X_train_balanced, y_train_balanced)
    
    y_pred = model.predict(X_test_preprocessed)
    y_prob = model.predict_proba(X_test_preprocessed)[:,1]  # for ROC-AUC
    
    print(f"--- {model.__class__.__name__} ---")
    print(classification_report(y_test, y_pred))
    print("ROC-AUC:", roc_auc_score(y_test, y_prob))
    print("\n")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV

gradient_boosting_params = {
    'loss':['log_loss', 'exponential'],
    'learning_rate':[0.01, 0.1,0.05],
    'n_estimators':[100,200,350],
    'max_depth': [3, 5, 7],
    'criterion':['friedman_mse', 'squared_error'],
    'max_features':['sqrt', 'log2']
}
```

```python
gb_random_search = RandomizedSearchCV(
    GradientBoostingClassifier(),
    param_distributions=gradient_boosting_params,
    n_iter=20,
    scoring='f1',
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    n_jobs=-1,
    verbose=2,
    random_state=42
)
```

```python
gb_random_search.fit(X_train_balanced, y_train_balanced)
```

```python
gb_random_search.best_params_
```

```python
from sklearn.metrics import accuracy_score


y_pred = gb_random_search.predict(X_test_preprocessed)
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, gb_random_search.predict_proba(X_test_preprocessed)[:,1]))
```
